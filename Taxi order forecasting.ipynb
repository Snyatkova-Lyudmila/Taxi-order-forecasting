### Анализ данных

#### Временной ряд

Прежде всего, для лучшего понимания данных изучим график временного ряда. Будем расматривать график временного ряда, ресемплированного по часам (данные data_1h), на который наложим графики: скользящее среднее за сутки, скользящее среднее за неделю.

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(15,8)
data_1h.plot(ax=ax, label='Исходные данные')
data_1h.rolling(24).mean().plot(ax=ax, c='red', label='Скользящее среднее / сутки')
data_1h.rolling(168).mean().plot(ax=ax, c='yellow', label='Скользящее среднее / неделя')
plt.title('График временного ряда, ресемплированного по 1 часу + скользящее среднее за неделю, сутки', fontsize=14)
plt.ylabel('Количество заказов')
plt.legend()
plt.show()
```

Вынесем на отдельные графики:

```python
# скользящее среднее за сутки + суточное стандартное отклонение
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(15,8)
data_1h.plot(ax=ax, label='Исходные данные')
data_1h.rolling(24).mean().plot(ax=ax, c='red', label='Скользящее среднее / сутки')
data_1h.rolling(24).std().plot(ax=ax, c='black', label='Стандартное отклонение')
plt.title('Cкользящее среднее за сутки + суточное стандартное отклонение', fontsize=14)
plt.ylabel('Количество заказов')
plt.legend()
plt.show()

# скользящее среднее за неделю + недельное стандартное отклонение
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(15,8)
data_1h.plot(ax=ax, label='Исходные данные')
data_1h.rolling(168).mean().plot(ax=ax, c='yellow', label='Скользящее среднее / неделя')
data_1h.rolling(168).std().plot(ax=ax, c='black', label='Стандартное отклонение')
plt.title('Cкользящее среднее за неделю + недельное стандартное отклонение', fontsize=14)
plt.ylabel('Количество заказов')
plt.legend()
plt.show()
```

#### Декомпозиция временного ряда

Разделим данные на три составляющие:  
* *тренд* (плавное изменение среднего значения без повторяющихся закономерностей);
* *сезонность* (циклично повторяющиеся закономерности);
* *остаток* (случайные изменения, не связанные с трендами или сезонностью).

**Построим графики декомпозиции** ***(на данных, ресемплированных по одному часу - data_1h)***

```python
decomposed_1h = seasonal_decompose(data_1h)

plt.figure(figsize=(15, 15))
plt.subplot(311)
decomposed_1h.trend.plot(ax=plt.gca(), grid=True)
plt.title('Trend', fontsize=14)
plt.subplot(312)
decomposed_1h.seasonal.plot(ax=plt.gca(), grid=True)
plt.title('Seasonality', fontsize=14)
plt.subplot(313)
decomposed_1h.resid.plot(ax=plt.gca(), grid=True)
plt.title('Residuals', fontsize=14)

plt.tight_layout()
```

* **выведем отдельно месячную и недельную сезонность** ***(на данных, ресемплированных по одному часу - data_1h)*** 

Так как на текущем графике проанализировать сезонность сложно, сделаем срез за один месяц и одну неделю (возьмем для рассмотрения, к примеру, май и одну неделю в середине мая: 14-20 мая 2018).

```python
plt.figure(figsize=(15, 5))
plt.suptitle('Месячная и недельная сезонность (data_1h)', fontsize=15)

plt.subplot(121)
decomposed_1h.seasonal['2018-05-01':'2018-05-31'].plot(ax=plt.gca(), grid=True)
plt.title('Seasonality 1 month / май 2018', fontsize=13)
plt.subplot(122)
decomposed_1h.seasonal['2018-05-14':'2018-05-20'].plot(ax=plt.gca(), grid=True)
plt.title('Seasonality 1 week / 14-20 мая 2018', fontsize=13)
plt.tight_layout()
```

***Trend*** 

На всем рассматриваемом периоде наблюдаем тренд на увеличение количества заказов. На графиках можно заметить, что в 2018 г. спрос на такси с начала марта по конец августа увеличился ~ в 5 раз (в начале марте количество заказов такси ~30 в час / ~750 в день, в конце августе ~155 в час / ~3500 в день)

***Seasonality*** 

Наблюдаем *недельную* и *суточную* сезонность:

* реже всего такси заказывают по вторникам и воскресеньям, чаще всего - по понедельникам и пятницам;  
* пик активности заказа такси наблюдаем до 03:00 и около 00:00, самый низкий спрос - около 06:00, также такси начинают активнее заказывать в часы пик (около 09:00, 18:00).

***Residuals*** 

Сильные колебания на графике остатков вероятно указывают на наличие пиковых значений.

#### Проверка стационарности ряда с помощью теста Дики-Фуллера

Как уже было отмечено ранее, со временем увеличивается дисперсия ряда. Присутствует тренд: количество заказов такси в аэропортах увеличивается со временем. Также в данных была выявлена сезонность (недельная и суточная).

Так как со временем меняется среднее и стандартное отклонение, то можно говорить о том, что такой ряд не является стационарным. Подтвердим это утверждение с помощью критерия Дики-Фуллера.

Будем проверять следующие *гипотезы*:

* *нулевая гипотеза* - временной ряд является нестацинарным;
* *альтернативная гипотеза* - временной ряд стационарен

```python
# зададим уровень значимости
alpha1 = 0.01
alpha2 = 0.05 

result = adfuller(data_1h['num_orders'])
p_value = result[1]
print('p_value =', p_value)
print('------------------------------------------------------------------------------------')
print('ADF Statistic:', result[0])
print('Critical Values:', result[4])
```

```python
print('Для уровня статистической значимости 5%')
if p_value < alpha2:
    print('Нулевая гипотеза отвергается: по критерию Дики-Фуллера ряд стационарен')
else:
    print('Нулевая гипотеза не отвергается: по критерию Дики-Фуллера ряд нестационарен')
```

```python
print('Для уровня статистической значимости 1%')
if p_value < alpha1:
    print('Нулевая гипотеза отвергается: по критерию Дики-Фуллера ряд стационарен')
else:
    print('Нулевая гипотеза не отвергается: по критерию Дики-Фуллера ряд нестационарен')
```

Таким образом, имеем пограничный результат: для уровня статистической значимости 5% по критерию Дики-Фуллера ряд стационарен; для уровня статистической значимости 1% по критерию Дики-Фуллера ряд нестационарен.

Можно говорить о ложном срабатывании теста - дело в том, что в данных есть тренд и сезонность, а это "по определению" исключает стационарность. Но для технологии, которую мы будем использовать в этом проекте (построение признаков и обучение модели машинного обучения) стационарность не является обязательной.

### Добавление признаков

```python
data_new = data_1h.copy()
```

* В данных была обнаружена недельная сезонность, таким образом необходимо добавить *календарный признак ('день недели')*.

```python
# добавим признак 'день недели', который отнесем к категориальному типу
data_new['dayofweek'] = data_new.index.dayofweek
data_new['dayofweek'] = data_new['dayofweek'].astype('category')
```

* Также стоит добавить *признак 'час'*, поскольку в данных есть суточная сезонность и информация о часе может быть полезной.

```python
# добавим признак 'час', который отнесем к категориальному типу
data_new['hour'] = data_new.index.hour
data_new['hour'] = data_new['hour'].astype('category')
```

---
Сезонность внутри месяца мы не обнаружили, следовательно, признак 'день' рассматривать не будем.

Также не будем добавлять к признакам 'месяц', поскольку данных у нас немного, всего 6 месяцев, при этом данные тренда мы учтем с помощью скользящего среднего и отстающих значений.

* Добавим *значения за предыдущее время (LAG)*

Ранее мы обозначили MAX_LAG = 168. Поскольку мы рассматриваем данные, ресемплированные по 1 часу, а также была замечена сезонность в течение недели, будет корректно рассматреть данные как минимум за 7 дней (24 * 7 = 168)

```python
for lag in range(1, MAX_LAG+1):
    data_new['lag_{}'.format(lag)] = data_new['num_orders'].shift(lag) 
data_new.dropna(inplace=True)
```

*  Добавим *признак 'cкользящее среднее'*

Скользящее среднее отвечает за тренд. И если взять скользящее среднее равное неделе, то этот трен будет гораздо болеее "ровный".

А сезонности у нас две, суточная и недельная. Поэтому оптимальнее использовать скользящее среднее за 168 отстающих значений  
(в начале проекта в разделе объявления констат обозначили ROLLING_MEAN_SIZE = 168).

```python
data_new['rolling_mean'] = data_new['num_orders'].shift().rolling(ROLLING_MEAN_SIZE).mean()
data_new.dropna(inplace=True)
```

Все полученные результаты (модель, рассчитанную метрику RMSE для каждой модели, время обучения) будем сохранять в массив result_models.

```python
result_models = []
```

```python
tscv = TimeSeriesSplit(n_splits=3)
```

>Для кроссвалидации и подбора гиперпараметров модели создадим функцию **best_model**:  

```python
def best_model(model, params, X, y, randomized_search=False, cv=tscv, vb=False):
    
    start_time = time.time()
    if randomized_search:
        grid = RandomizedSearchCV(model,
                                  params, 
                                  scoring='neg_root_mean_squared_error',
                                  cv=cv,
                                  n_jobs=-1,
                                  verbose=vb,
                                  random_state=RANDOM_STATE)
    else:
        grid = GridSearchCV(model,
                            params, 
                            scoring='neg_root_mean_squared_error',
                            cv=cv,
                            n_jobs=-1,
                            verbose=vb)
    
    grid.fit(X, y)
    best_score = round(grid.best_score_ * (-1), 2)
    learninig_time = round((time.time() - start_time) / 60, 2) 
    
    return grid, best_score, learninig_time
```

```python
# создадим списки с названиями признаков
cat_col_names = X_train.select_dtypes(include='category').columns.to_list()
num_col_names = X_train.select_dtypes(exclude='category').columns.to_list()
```

```python
# преобразование признаков
ohe_col_transf = make_column_transformer( 
    (OneHotEncoder(drop='first', handle_unknown='ignore'), cat_col_names),
    (StandardScaler(), num_col_names), 
    remainder='passthrough',
    verbose_feature_names_out=False)

ord_col_transf = make_column_transformer( 
    (OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_col_names),
    (StandardScaler(), num_col_names),
    remainder='passthrough')
```
